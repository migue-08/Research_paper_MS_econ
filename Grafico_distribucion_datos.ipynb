```python
import numpy as np
import os
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
plt.rc("figure", figsize=(10, 6))
np.set_printoptions(precision=4)
pd.options.display.max_columns = 20
pd.options.display.max_rows = 20
pd.options.display.max_colwidth = 80
```

# Load data

## Saber 11 - Total

```python
chunker = pd.read_csv("Resultados__nicos_Saber_11_20250123.csv", chunksize=1184950, low_memory=False)
```

```python
tot = pd.Series([], dtype='int64')
for piece in chunker:
    piece["año"] = piece["PERIODO"].astype(str).str[:4]
    piece = piece.drop_duplicates()
    tot = tot.add(piece["año"].value_counts(), fill_value=0)

tot = tot.sort_values(ascending=False)
```

## Graph

```python
sns.set_style('ticks')
sns.barplot(tot,
            color='gray')
plt.title('Pruebas Icfes Saber11')
plt.xlabel(None)
plt.ylabel('Estudiantes')
plt.ticklabel_format(axis='y', style='plain')

# plt.savefig('info_por_ano.pdf', bbox_inches='tight')
```
